## Environments

Python is a very popular programming language in life science. Therefore there are many different libraries and packages that allow you to perform many common tasks:

- [pandas](https://pandas.pydata.org/): data manipulation and analysis
- [matplotlib](https://matplotlib.org/): data visualization
- [scikit-image](https://scikit-image.org/): image processing
- ...

These packages develop very fast and many different versions of them exist. Depending on the complexity of your experiments, sometimes you will need to use very specific versions of a library/package. But if you install those versions, other packages that you have installed may stop working. To obviate this issue you can create "environments". 

Having different environments on your computer is as if you would have different computers, that don't really talk to each other, which is good, because then you can have different versions of different libraries and make the best use of all of them, without running into unpleasant incompatibility problems between different versions of different packages and unsolvable dependencies issues. You could even have different python versions running on your computer if you enclose them in different environments.

**To use napari we want to build a specific environment that will contain it!**

## Create a new environment
Now that you have installed mamba, and you have added it to your PATH, you can start making new environments for your projects. First of all open a new command prompt(Windows) or a terminal (macOS, Linux).

````python
Image('CommandPrompt.png')
````

### Anatomy of a terminal
As default, each time you open a new terminal you will be located in the **"base" environment**.<br>
The name of the environment you are in appears in round brackets on the left of a new command line. But sometimes this is not the case as in the picture above, however, if nothing is specified, remember that you are working in the base environment.<br>

What follows is the path of the directory that you are working in. For example, all the outputs of specific operations will be saved in this directory (in the example "C:\Users\anast").

**NEVER USE YOUR BASE ENVIRONMENT!!!**

For every new project, you should create a new environment. And this is what we will do now for napari. We will create a new environment with the python version that we want.

## Create a new environment for your napari project
You can name your environment as you wish. In this example we will call it **"napari-env"**.

Type in the terminal the following code:

````python
mamba create -y -n napari-env -c conda-forge python=3.9
````

"mamba create" is your command, what follows are some specifications:

- -y means "yes": Basically you confirm that you want to create this environment, if you don't add this "-y" at this stage, at some point you will be asked if you are sure that you want to create the environment and you will need to confirm by typing Y. In order to skip this step we confirm it upfront
- -n means "name": what follow this n will be used as name of the new environment you are creating, here you can be creative
- "napari-env": is the name we just decided to assign to our environment/project
- -c means "channel": all these packages and libraries are stored somewhere on the internet, in order to tell your computer where it has to go and download them, you have to specify this variable channel.
- The most common channel from which to download libraries is ["conda-forge"](https://conda-forge.org/#page-top). Another popular repository for libraries and packages is [**PyPi**](https://pypi.org/), but to use it you will see the specific command **pip install**
- python=3.9: of course, we can create an environment that it is completely empty, and just has a name, but this will be rather unuseful, therefore as first step when we create a new environment we also install python in it. In this case, we have decided to go for the 3.9 version of python stored on conda-forge

If the code runs correctly you will be able to enter in the newly created environment as follow:

````python
mamba activate napari-env
````

````python
Image('new_env_windows.png')
````

Now you see that you have entered the new environment that you have created, and this is specified in the round brackets as **napari-env**.<br>
But so far in your environment, you have only installed python and nothing else, so in order to use napari, you need to install it in this specific environment.

```warnings
Make always sure that you have activated the correct environment before installing anything in it!!!<br>
Be careful, if nothing is specified in brackets it means that you are in the base environment.
```



## Install napari
After activating the correct environment, you need to install napari in it by using this code:

````python
mamba install -c conda-forge napari pyqt
````

Here you are asking mamba to install **napari** and **pyqt** from conda-forge.<br>
pyqt is an additional essential package to run napari. In this line you run the installation of the two in one go, but if you understand how the principle of command lines works now, you cound have installed them one after the other.

If napari is installed successfully in your new environment, you should be able to open the napari viewer by typing:

````python
napari
````

Sometimes you may need to restart the kernel to use updated packages. In this case use:

````python
mamba deactivate
````

To go back in the base environment, and then:

````python
mamba activate napari-env
````

````python
mamba env list
````

## Install napari plugins

### 1. Use the napari installer
Certainly the easiest way to install a plugin for napari is to use the napari installer built into the software itself. Any plugin that is uploaded on Pypi is automatically listed in the [napari hub](https://www.napari-hub.org/).

In this example we will install the [napari-segment-blobs-and-things-with-membranes](https://github.com/haesleinhuepf/napari-segment-blobs-and-things-with-membranes) plugin developed by Rober Haase.

### 2. Install plugins from command line
The plugins can be also installed directly from the command line. As most of them are on Pypi the most direct way to install them is to use "pip install".

Make sure that you have activated the environment where your napari is installed, and run the following command line.

````python
pip install napari-segment-blobs-and-things-with-membranes
````

After installing a plugin like this, next time you open napari, the plugin should be listed in your plugins.

###  3. Install plugins from github
Some of the plugins are not stable enough to be on pip yet, but you may want to use them. In this case you can download them directly from the repositories where they are being developed on GitHub. This is the case for example of our napari-morphodynamics.<br>

If you would like to use it you could install it with the following command:
pip install **git+** URL to the repository you wish to install **.git**

````python
pip install git+https://github.com/guiwitz/napari-morphodynamics.git
````

This method can be used for any library or package. But be aware that when you install them from GitHub they might not be stable and contain bugs.
